# 4. Análise Exploratória dos Dados

Agora que conhecemos os dados, vamos começar com uma análise exploratória de todos os dados que temos disponíveis.

Usaremos a seguinte função para analisar cada indicador economico hh

```
def exploratory_analysis(indicator, column=None):
    df = indicator.processed_storage.get_data()

    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    column = column or numeric_cols[0]
    data = df[column].dropna().tolist()
    stats = descriptive.summary(data)

    print(f"Resumo Estatístico - '{column}':")

    for key, value in stats.items():
        print(f"  {key:12s}: {value:.4f}")
```
