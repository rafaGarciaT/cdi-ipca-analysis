# 0. Introdução

Esse é o primeiro Notebook do projeto de armazenamento e análise de indicadores econômicos brasileiros.
Neles, comentaremos sobre os dados coletados e faremos uma análise completa e, no final, teremos insights sobre como o mercado brasileiro funciona.

## 0.1 Objetivo

1. Criar uma ferramenta que junta os principais indicadores em um lugar só de maneira eficiente e precisa.
2. Facilitar tanto o aprendizado de como o mercado brasileiro funciona, quanto insights adicionais, a partir de notebooks de análise e dashboards.

## 0.2 Escopo

O escopo desse projeto envolve, na parte executável, o armazenamento das taxas dos principais indicadores econômicos a partir de APIs públicas, o processamento desses dados, que envolve o cálculo de todas as mensurações em cada indicador, e o armazenamento desses dados.
Adicionalmente, o escopo do projeto também envolve a parte de análise de dados, com Notebooks e Dashboards exploratórios.

## 0.3 Estrutura do Projeto

A extração dos dados acontece inteiramente em src/, e é orquestrada por src/pipeline.py. As subpastas de src/ auxiliam separam as funções chamadas na pipeline.
- **src/fetch**: Lógica de requisição de APIs. O request para cada fonte é reunido em src/fetch/base.py, e implementado nos módulos de fetch referentes.
- **src/transform**: Lógica do cálculo de mensurações para cada indicador.
- **src/storage**: Lógica do armazenamento dos dados coletados. Separado entre processed e raw. base.py de cada subpasta define a estrutura que cada classe de storage deve seguir.
- **src/indicators**: Lógica de requisição, tratamento e armazenamento de dados brutos e processados implementada da maneira necessária para cada indicador econômico. O objeto Indicator é o que a pipeline vai chamar.
- Outras subpastas são para módulos auxiliares para o projeto.
- config.py é um módulo de constantes de configuração utilizados no projeto.

A Pipeline tem três modos de funcionamento. Monthly, onde processamos a data do mês atual. Yearly, onde processamos todos os meses de um ano específico. E backfill, onde as lacunas entre os valores armazenados são preenchidos.
Por enquanto, a Pipeline só armazena valores lançados mensalmente. Taxas acumuladas no ano são calculadas em cada mês processado, e o valor acumulado de um ano inteiro estará presente no mês de Dezembro daquele ano.
